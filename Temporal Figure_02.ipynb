```python
import numpy as np
import matplotlib.pyplot as plt  # To visualize
import pandas as pd  # To read data
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_excel('Data For Temporal Plot.xlsx')  # load data set
Month = data.iloc[:, 0].values.reshape(-1, 1)  # values converts it into a numpy array
CDD = data.iloc[:, 1].values.reshape(-1, 1)  # -1 means that calculate the dimension of rows, but have 1 column
linear_regressor = LinearRegression()  # create object for the class
linear_regressor.fit(Month, CDD)  # perform linear regression
Y_pred = linear_regressor.predict(Month)  # make predictions
```

```python
plt.plot(Month, CDD)
plt.plot(Month, Y_pred, color='red')
plt.show()
```
