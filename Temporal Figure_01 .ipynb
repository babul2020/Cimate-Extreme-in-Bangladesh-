```python
#importing the libaries 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt # To visualize
from sklearn.linear_model import LinearRegression
from scipy import stats
import operator as op 
```

```python
#Importing data
df = pd.read_excel('Data For Temporal Plot.xlsx',sheet_name = 'Sheet2')

#temperature.head()
df
```

```python
Month = df.iloc[0:11, 0].values.reshape(-2, 1)  # values converts it into a numpy array
Rx5day_a = df.iloc[0:11, 3].values.reshape(-2, 1)  # -1 means that calculate the dimension of rows, but have 1 column

linear_regressor = LinearRegression()  # create object for the class
linear_regressor.fit(Month, Rx5day_a)  # perform linear regression
Y_pred = linear_regressor.predict(Month)  # make predictions
```

```python
plt.plot( df['Month'], df[""], color='red', marker='o')
#plt.plot(Month, CDD)
plt.plot(Month, Y_pred, color='blue')

plt.show()
```

```python
#Viewing the data

plt.plot( df['Month'], df ['DRT_a'], color='red', marker='o')
plt.title('Month vs CDD', fontsize=14)
plt.xlabel('Month', fontsize=14)
plt.ylabel('CDD', fontsize=14)
plt.plot(Month, Y_pred, color='blue')
plt.grid(True)
plt.show()

# # calculate Slope, R_Squre, P_value

# slope, intercept, r_value, p_value, std_err = stats.linregress(df['Month'], df['CDD'])
# df['linreg'] = intercept + slope * df.Month  

# print('Slope: ',slope,'\nR Squre:', r_value**2, '\nP_value: ', p_value) 
```

```python
#Mean_CDD Calculation 

df["CDD"].mean()

#ctrl + enter 

# Anomoly_CDD detection 

CDD = df.CDD
anomoly = op.sub(CDD, df["CDD"].mean())
print(anomoly)
```

```python
# calculate Slope, R_Squre, P_value

slope, intercept, r_value, p_value, std_err = stats.linregress(df['Month'], df['CDD'])
df['linreg'] = intercept + slope * df.Month  

print('Slope: ',slope,'\nR Squre:', r_value**2, '\nP_value: ', p_value) 
```
